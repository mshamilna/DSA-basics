### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

#### 5.a Pie chart on distribution of classifications in the dataset.

```python
plt.pie(df.Classification.value_counts(), 
        labels = df.Classification.unique(),
        autopct='%1.1f%%', 
        startangle=90
       ) 
plt.legend(title='Classification')
plt.title('Classification')
plt.show()
```

#### 5.b Bar chart on 

```python
plt.bar(df.Classification, df.SL, color ='r', width = 0.25, label ='SL')
plt.bar(df.Classification, df.SW, color ='g', width = 0.25, label ='SW')
plt.bar(df.Classification, df.PL, color ='y', width = 0.25, label ='PL')
plt.bar(df.Classification, df.PW, color ='b', width = 0.25, label ='PW')
plt.show()
```

#### 5.c Scatter diagram

```python
plt.scatter(df.Classification,df.SL)
plt.scatter(df.Classification,df.SW)
plt.scatter(df.Classification,df.PL)
plt.scatter(df.Classification,df.PW)
plt.show()
```

#### 5.d Line diagram

```python
plt.plot(df.SL, label='SL')
plt.plot(df.SW, label='SW')
plt.plot(df.PL, label='PL')
plt.plot(df.PW, label='PW')
plt.title('DATA FRAME')
plt.xlabel("Count")
plt.ylabel("Value")
plt.legend()
plt.show()
```
